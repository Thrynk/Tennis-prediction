```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(os.path.abspath('../input/diff-cleaned-featured-2008-2017.csv'))
```

```python
df.columns.values
```

We want to standardize features so that the algorithm doesn't give more importance to a feature :

Let's check the dstirbutions of our features :

```python
fig, axes = plt.subplots(nrows=7, ncols=2, figsize=(8,10))
plt.subplots_adjust(top=2)
plt.subplot(7, 2, 1)
sns.distplot(df['direct'])
plt.subplot(7, 2, 2)
sns.distplot(df['fs_diff'])
plt.subplot(7, 2, 3)
sns.distplot(df['w1sp_diff'])
plt.subplot(7, 2, 4)
sns.distplot(df['w2sp_diff'])
plt.subplot(7, 2, 5)
sns.distplot(df['wrp_diff'])
plt.subplot(7, 2, 6)
sns.distplot(df['tpw_diff'])
plt.subplot(7, 2, 7)
sns.distplot(df['tmw_diff'])
plt.subplot(7, 2, 8)
sns.distplot(df['aces_per_game_diff'])
plt.subplot(7, 2, 9)
sns.distplot(df['df_per_game_diff'])
plt.subplot(7, 2, 10)
sns.distplot(df['bp_won_diff'])
plt.subplot(7, 2, 11)
sns.distplot(df['wsp_diff'])
plt.subplot(7, 2, 12)
sns.distplot(df['completeness_diff'])
plt.subplot(7, 2, 13)
sns.distplot(df['serve_adv_diff'])
plt.subplot(7,2,14)
sns.distplot(df['elo_rating_diff'])
plt.show()
```

```python
df['fs_diff'] = df['fs_diff'] / df['fs_diff'].std()
df['w1sp_diff'] = df['w1sp_diff'] / df['w1sp_diff'].std()
df['w2sp_diff'] = df['w2sp_diff'] / df['w2sp_diff'].std()
df['wrp_diff'] = df['wrp_diff'] / df['wrp_diff'].std()
df['tpw_diff'] = df['tpw_diff'] / df['tpw_diff'].std()
df['tmw_diff'] = df['tmw_diff'] / df['tmw_diff'].std()
df['aces_per_game_diff'] = df['aces_per_game_diff'] / df['aces_per_game_diff'].std()
df['df_per_game_diff'] = df['df_per_game_diff'] / df['df_per_game_diff'].std()
df['bp_won_diff'] = df['bp_won_diff'] / df['bp_won_diff'].std()
df['wsp_diff'] = df['wsp_diff'] / df['wsp_diff'].std()
df['completeness_diff'] = df['completeness_diff'] / df['completeness_diff'].std()
df['serve_adv_diff'] = df['serve_adv_diff'] / df['serve_adv_diff'].std()
df['elo_rating_diff'] = df['elo_rating_diff'] / df['elo_rating_diff'].std()
```

```python
fig, axes = plt.subplots(nrows=7, ncols=2, figsize=(8,10))
plt.subplots_adjust(top=2)
plt.subplot(7, 2, 1)
sns.distplot(df['direct'])
plt.subplot(7, 2, 2)
sns.distplot(df['fs_diff'])
plt.subplot(7, 2, 3)
sns.distplot(df['w1sp_diff'])
plt.subplot(7, 2, 4)
sns.distplot(df['w2sp_diff'])
plt.subplot(7, 2, 5)
sns.distplot(df['wrp_diff'])
plt.subplot(7, 2, 6)
sns.distplot(df['tpw_diff'])
plt.subplot(7, 2, 7)
sns.distplot(df['tmw_diff'])
plt.subplot(7, 2, 8)
sns.distplot(df['aces_per_game_diff'])
plt.subplot(7, 2, 9)
sns.distplot(df['df_per_game_diff'])
plt.subplot(7, 2, 10)
sns.distplot(df['bp_won_diff'])
plt.subplot(7, 2, 11)
sns.distplot(df['wsp_diff'])
plt.subplot(7, 2, 12)
sns.distplot(df['completeness_diff'])
plt.subplot(7, 2, 13)
sns.distplot(df['serve_adv_diff'])
plt.subplot(7,2,14)
sns.distplot(df['elo_rating_diff'])
plt.show()
```
