```python
import pandas as pd
import os
```

```python
train_df = pd.read_csv(os.path.abspath('../input/cleaned-featured-2008-2017.csv'))
```

```python
train_df.info()
```

```python
train_df['fs_diff'] = train_df['p_fs'] - train_df['o_fs']
train_df['w1sp_diff'] = train_df['p_w1sp'] - train_df['o_w1sp']
train_df['w2sp_diff'] = train_df['p_w2sp'] - train_df['o_w2sp']
train_df['wrp_diff'] = train_df['p_wrp'] - train_df['o_wrp']
train_df['tpw_diff'] = train_df['p_tpw'] - train_df['o_tpw']
train_df['tmw_diff'] = train_df['p_tmw'] - train_df['o_tmw']
train_df['aces_per_game_diff'] = train_df['p_aces_per_game'] - train_df['o_aces_per_game']
train_df['df_per_game_diff'] = train_df['p_df_per_game'] - train_df['o_df_per_game']
train_df['bp_won_diff'] = train_df['p_bp_won'] - train_df['o_bp_won']
train_df['wsp_diff'] = train_df['p_wsp'] - train_df['o_wsp']
train_df['completeness_diff'] = train_df['p_completeness'] - train_df['o_completeness']
train_df['serve_adv_diff'] = train_df['p_serve_adv'] - train_df['o_serve_adv']
```

```python
train_df.head()
```

```python
train_df = train_df.drop(['p_fs', 'o_fs', 'p_w1sp','o_w1sp','p_w2sp','o_w2sp','p_wrp','o_wrp','p_tpw','o_tpw','p_tmw','o_tmw','p_aces_per_game','o_aces_per_game'
               ,'p_df_per_game','o_df_per_game','p_bp_won','o_bp_won','p_wsp','o_wsp','p_completeness','o_completeness','p_serve_adv','o_serve_adv'], axis = 1)
```

```python
train_df.to_csv(os.path.abspath('../input/diff-cleaned-featured-2008-2017.csv'), index=False)
```
